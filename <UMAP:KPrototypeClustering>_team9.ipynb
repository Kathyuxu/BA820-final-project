## **Data Preprocessing**

Duplicates

```python
raw_data_no_duplicates = raw_data.loc[raw_data.groupby('id')['calendar_last_scraped'].idxmax()]
raw_data_no_duplicates['id'].duplicated().sum()
```

Missing Value

price

```python
# deal with NA in "price" --> drop all the missing value
raw_data_no_na_price = raw_data_no_duplicates.dropna(subset = ['price'], axis = 0)

# check the number of NA
raw_data_no_na_price['price'].isna().sum()
```

Host Response Rate and Host Acceptance Rate

```python
raw_data_no_na_rate = raw_data_no_na_price.copy()

# Remove '%' from rate columns and convert to float
raw_data_no_na_rate["host_response_rate"] = raw_data_no_na_rate['host_response_rate'].replace('[\%]', '', regex=True).astype(float)
raw_data_no_na_rate["host_acceptance_rate"] = raw_data_no_na_rate['host_acceptance_rate'].replace('[\%]', '', regex=True).astype(float)

# deal with NA in "host_acceptance_rate" and "host_response_rate" --> use mean to fill the NA
raw_data_no_na_rate["host_response_rate"] = raw_data_no_na_rate["host_response_rate"].fillna(raw_data_no_na_rate["host_response_rate"].mean())
raw_data_no_na_rate["host_acceptance_rate"] = raw_data_no_na_rate["host_acceptance_rate"].fillna(raw_data_no_na_rate["host_acceptance_rate"].mean())

# check the number of NA
print(raw_data_no_na_rate["host_response_rate"].isna().sum())
print(raw_data_no_na_rate["host_acceptance_rate"].isna().sum())
```

Bathrooms

```python
# deal with NA in "bathrooms" --> drop all the missing value
raw_data_no_na_bathroom = raw_data_no_na_rate.dropna(subset = ['bathrooms'], axis = 0)

# check the number of NA
print(raw_data_no_na_bathroom["bathrooms"].isna().sum())
```

Bedrooms

```python
# deal with NA in "bedrooms" --> use mean to fill the NA
raw_data_no_na_bedrooms = raw_data_no_na_bathroom.copy()
raw_data_no_na_bedrooms["bedrooms"] = raw_data_no_na_bedrooms["bedrooms"].fillna(raw_data_no_na_bedrooms["bedrooms"].mean())

# check the number of NA
print(raw_data_no_na_bedrooms["bedrooms"].isna().sum())
```

Beds

```python
# deal with NA in "bed" --> use function "# of bedrooms * avg(beds)" to fill the NA
raw_data_no_na_bed = raw_data_no_na_bedrooms.copy()
# calculate avgerage bed per room
avg_beds_per_bed = raw_data_no_na_bed["beds"].sum() / raw_data_no_na_bed["bedrooms"].sum()
# use function to fill the NA
function = raw_data_no_na_bed["bedrooms"] * avg_beds_per_bed
raw_data_no_na_bed["beds"] = raw_data_no_na_bed["beds"].fillna(function)

# check the number of NA
print(raw_data_no_na_bed["beds"].isna().sum())
```

Review Score

```python
# deal with NA in review score --> use mean to fill the NA
raw_data_no_na_review = raw_data_no_na_bed.copy()
raw_data_no_na_review["review_scores_rating"] = raw_data_no_na_review["review_scores_rating"].fillna(raw_data_no_na_review["review_scores_rating"].mean())
raw_data_no_na_review["review_scores_accuracy"] = raw_data_no_na_review["review_scores_accuracy"].fillna(raw_data_no_na_review["review_scores_accuracy"].mean())
raw_data_no_na_review["review_scores_cleanliness"] = raw_data_no_na_review["review_scores_cleanliness"].fillna(raw_data_no_na_review["review_scores_cleanliness"].mean())
raw_data_no_na_review["review_scores_checkin"] = raw_data_no_na_review["review_scores_checkin"].fillna(raw_data_no_na_review["review_scores_checkin"].mean())
raw_data_no_na_review["review_scores_communication"] = raw_data_no_na_review["review_scores_communication"].fillna(raw_data_no_na_review["review_scores_communication"].mean())
raw_data_no_na_review["review_scores_location"] = raw_data_no_na_review["review_scores_location"].fillna(raw_data_no_na_review["review_scores_location"].mean())
raw_data_no_na_review["review_scores_value"] = raw_data_no_na_review["review_scores_value"].fillna(raw_data_no_na_review["review_scores_value"].mean())

# check the number of NA
print(raw_data_no_na_review["review_scores_rating"].isna().sum())
print(raw_data_no_na_review["review_scores_accuracy"].isna().sum())
print(raw_data_no_na_review["review_scores_cleanliness"].isna().sum())
print(raw_data_no_na_review["review_scores_checkin"].isna().sum())
print(raw_data_no_na_review["review_scores_communication"].isna().sum())
print(raw_data_no_na_review["review_scores_location"].isna().sum())
print(raw_data_no_na_review["review_scores_value"].isna().sum())
```

```python
import umap
import plotly.express as px
# Use UMAP for dimentionality Reduction
reducer = umap.UMAP(n_components=3)
umap_prototype_embeddings = reducer.fit_transform(data_roomclass_k)

# Change the prototype_embeddings into dataframe
umap_prototype_df = pd.DataFrame(umap_prototype_embeddings, columns=["UMAP1", "UMAP2", "UMAP3"])
```

```python
# Re-add categorical columns
umap_prototype_df[cat_cols] = data_roomclass_k[cat_cols].reset_index(drop=True)

# Print columns to confirm fix
print("Updated UMAP DataFrame Columns:", umap_prototype_df.columns)
```

```python
from kmodes.kprototypes import KPrototypes
import numpy as np
import matplotlib.pyplot as plt

# Convert DataFrame to NumPy Array
data_array = umap_prototype_df.to_numpy()

# Identify Categorical Column Indices
categorical_indices = [umap_prototype_df.columns.get_loc(col) for col in cat_cols]

# Find the optimal number of clusters using the Elbow Method
costs = []
k_values = range(2, 10)

for k in k_values:
    kproto = KPrototypes(n_clusters=k, init='Huang', verbose=0, random_state = 25, max_iter=10, n_jobs=-1)
    kproto.fit_predict(data_array, categorical=categorical_indices)
    costs.append(kproto.cost_)

plt.figure(figsize=(8, 5))
plt.plot(k_values, costs, marker='o', linestyle='dashed', color='b')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Cost Function')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Choose the optimal k and run the final K-prototype model
optimal_k = 3

kproto_final = KPrototypes(n_clusters=optimal_k, init='Huang', verbose=2,random_state = 42)
clusters = kproto_final.fit_predict(data_array, categorical=categorical_indices)

umap_prototype_df['Cluster_labels'] = clusters
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(umap_prototype_df[["UMAP1", "UMAP2", "UMAP3"]], clusters)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
visualization = data_roomclass_k.dropna(subset=['cluster_label'])

cols_to_check = ['property_type_numeric', 'room_type_numeric']
clusters = sorted(visualization['cluster_label'].unique())

room_type_mapping = {0: 'Entire home/apt', 1: 'Private room', 2: 'Shared room'}
property_type_mapping = {
    0: 'Standalone Homes',
    1: 'Apartments & Condos',
    2: 'Shared Spaces',
    3: 'Private Rooms',
    4: 'Hotels & Boutique Stays'
}

distribution_list = []
for cluster in clusters:
    cluster_data = visualization[visualization['cluster_label'] == cluster]
    for col in cols_to_check:
        value_counts = cluster_data[col].value_counts().reset_index()
        value_counts.columns = ['Value', 'Count']
        value_counts['Cluster'] = cluster
        value_counts['Category'] = col
        distribution_list.append(value_counts)

distribution_df = pd.concat(distribution_list, ignore_index=True)

fig, ax = plt.subplots(figsize=(12, 6))
property_data = distribution_df[distribution_df['Category'] == 'property_type_numeric'].copy()
property_data['Value'] = property_data['Value'].map(property_type_mapping)

sns.barplot(data=property_data, x='Value', y='Count', hue='Cluster', palette="Blues", edgecolor="black", ax=ax)

ax.set_title("Comparison of Property Type Distribution Across Clusters", fontsize=14)
ax.set_xlabel("Property Type")
ax.set_ylabel("Count")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.legend(title="Cluster")
plt.tight_layout()
plt.savefig('cluster_property_type_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

fig, ax = plt.subplots(figsize=(12, 6))
room_data = distribution_df[distribution_df['Category'] == 'room_type_numeric'].copy()
room_data['Value'] = room_data['Value'].map(room_type_mapping)

sns.barplot(data=room_data, x='Value', y='Count', hue='Cluster', palette="Blues", edgecolor="black", ax=ax)

ax.set_title("Comparison of Room Type Distribution Across Clusters", fontsize=14)
ax.set_xlabel("Room Type")
ax.set_ylabel("Count")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.legend(title="Cluster")
plt.tight_layout()
plt.savefig('cluster_room_type_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```
